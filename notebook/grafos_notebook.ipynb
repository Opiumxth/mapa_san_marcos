```python
import tkinter as tk
from tkinter import messagebox
import random

# Funci칩n para imprimir tablero en consola
def imprimir_tablero(tablero):
    for fila in range(9):
        print(" ".join(str(num) if num != 0 else "." for num in tablero[fila]))

# Verifica si es v치lido colocar un n칰mero
def es_valido(tablero, fila, col, num):
    for i in range(9):
        if tablero[fila][i] == num or tablero[i][col] == num:
            return False
    inicio_fila, inicio_col = (fila // 3) * 3, (col // 3) * 3
    for i in range(inicio_fila, inicio_fila + 3):
        for j in range(inicio_col, inicio_col + 3):
            if tablero[i][j] == num:
                return False
    return True

# Algoritmo de backtracking para resolver Sudoku
def resolver_sudoku(tablero):
    for fila in range(9):
        for col in range(9):
            if tablero[fila][col] == 0:
                for num in range(1, 10):
                    if es_valido(tablero, fila, col, num):
                        tablero[fila][col] = num
                        if resolver_sudoku(tablero):
                            return True
                        tablero[fila][col] = 0
                return False
    return True

# Generar tablero aleatorio simple (parcialmente lleno)
def generar_tablero_random():
    tablero = [[0 for _ in range(9)] for _ in range(9)]
    for _ in range(random.randint(10, 25)):  # Agrega entre 10 y 25 n칰meros
        fila, col = random.randint(0, 8), random.randint(0, 8)
        num = random.randint(1, 9)
        if tablero[fila][col] == 0 and es_valido(tablero, fila, col, num):
            tablero[fila][col] = num
    return tablero

# Interfaz gr치fica con Tkinter
class SudokuApp:
    def init(self, root):
        self.root = root
        self.root.title("游빌 Sudoku Interactivo")
        self.entradas = [[None for _ in range(9)] for _ in range(9)]
        self.crear_tablero()
        self.crear_botones()

    def crear_tablero(self):
        for fila in range(9):
            for col in range(9):
                entrada = tk.Entry(self.root, width=2, font=('Arial', 18), justify='center')
                entrada.grid(row=fila, column=col, padx=2, pady=2)
                self.entradas[fila][col] = entrada

    def crear_botones(self):
        tk.Button(self.root, text="Resolver", command=self.resolver, bg="lightgreen").grid(row=10, column=0, columnspan=3)
        tk.Button(self.root, text="Borrar", command=self.borrar_tablero, bg="lightgray").grid(row=10, column=3, columnspan=3)
        tk.Button(self.root, text="Aleatorio", command=self.generar_aleatorio, bg="lightblue").grid(row=10, column=6, columnspan=3)

    def obtener_tablero(self):
        tablero = []
        for fila in range(9):
            fila_valores = []
            for col in range(9):
                val = self.entradas[fila][col].get()
                if val.isdigit():
                    fila_valores.append(int(val))
                else:
                    fila_valores.append(0)
            tablero.append(fila_valores)
        return tablero

    def es_tablero_valido(self, tablero):
        for fila in range(9):
            for col in range(9):
                num = tablero[fila][col]
                if num != 0:
                    tablero[fila][col] = 0
                    if not es_valido(tablero, fila, col, num):
                        return False
                    tablero[fila][col] = num
        return True

    def resolver(self):
        tablero = self.obtener_tablero()
        copia = [fila[:] for fila in tablero]

        if not self.es_tablero_valido(tablero):
            messagebox.showerror("Error", "El tablero contiene errores o duplicados.")
            return

        if resolver_sudoku(copia):
            for fila in range(9):
                for col in range(9):
                    self.entradas[fila][col].delete(0, tk.END)
                    self.entradas[fila][col].insert(0, str(copia[fila][col]))
        else:
            messagebox.showerror("Sin soluci칩n", "El Sudoku no tiene soluci칩n.")

    def borrar_tablero(self):
        for fila in range(9):
            for col in range(9):
                self.entradas[fila][col].delete(0, tk.END)

    def generar_aleatorio(self):
        tablero = generar_tablero_random()
        self.borrar_tablero()
        for fila in range(9):
            for col in range(9):
                if tablero[fila][col] != 0:
                    self.entradas[fila][col].insert(0, str(tablero[fila][col]))

# Ejecutar la app
if name == "main":
    root = tk.Tk()
    app = SudokuApp(root)
    root.mainloop()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear grafo dirigido ponderado
G = nx.DiGraph()
G.add_weighted_edges_from([
    ('A', 'B', 4),
    ('A', 'C', 2),
    ('B', 'C', 1),
    ('B', 'D', 5),
    ('C', 'D', 8),
    ('C', 'E', 10),
    ('D', 'E', 2),
    ('E', 'F', 3),
    ('D', 'F', 6)
])

# Dijkstra desde nodo origen
origen = 'A'
destino = 'F'
camino = nx.dijkstra_path(G, origen, destino, weight='weight')
longitud = nx.dijkstra_path_length(G, origen, destino, weight='weight')

# Mostrar resultado
print(f"Camino m치s corto de {origen} a {destino}: {camino}")
print(f"Distancia total: {longitud}")

# Visualizaci칩n
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)})
path_edges = list(zip(camino, camino[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3)
plt.title("Ruta m치s corta usando Dijkstra")
plt.show()
```
